# EDA Diamonds

```python
#Imports:

import pandas as pd
import numpy as np
import seaborn as sbs
import matplotlib.pylab as plt


```

```python
city_df=pd.read_csv('data/csv_tables/diamonds_city.csv') #2 columns: city and city_id
clarity_df=pd.read_csv('data/csv_tables/diamonds_clarity.csv')#2 columns: clarity and clarity_id
color_df=pd.read_csv('data/csv_tables/diamonds_color.csv')#2 columns: color and color_id
cut_df=pd.read_csv('data/csv_tables/diamonds_cut.csv')#2 columns: cut and cut_id
dimensions_df=pd.read_csv('data/csv_tables/diamonds_dimensions.csv')#6 columns: depth, table, x, y z and index_id
properties_df=pd.read_csv('data/csv_tables/diamonds_properties.csv')#4 columns: index_id, cut_id, color_id, clarity_id
transactional_df=pd.read_csv('data/csv_tables/diamonds_transactional.csv')#4 columns: index_id, price, city_id, carat

transactional_df
```

```python
diamonds_df=pd.merge(city_df, transactional_df)
diamonds_df.drop('city_id', axis=1, inplace=True) #del axis 1 que son las columnas. Quito el city_id porque ya no lo necesito para nada.
#recordatorio: hacer inplace es para no tener que volver a definir el df, que te modifique sobre el que estás realizando las moficiaciones.
```

```python
diamonds_df=pd.merge(diamonds_df, properties_df)
diamonds_df= pd.merge(diamonds_df, dimensions_df)
diamonds_df=pd.merge(diamonds_df, cut_df)
diamonds_df=pd.merge(diamonds_df, color_df)
diamonds_df=pd.merge(diamonds_df,clarity_df)

```

```python
diamonds_df.drop('index_id', axis=1, inplace=True )
diamonds_df.drop('cut_id', axis=1, inplace=True )
diamonds_df.drop('color_id', axis=1, inplace=True )
diamonds_df.drop('clarity_id', axis=1, inplace=True )
```

```python
diamonds_df
```

```python
mean_price = diamonds_df['price'].mean()
max_price=diamonds_df['price'].max()
min_price=diamonds_df['price'].min()
print('Price Mean:', mean_price)
print('Max price:', max_price)
print('Min price:', min_price)
```

```python
median_price = diamonds_df['price'].median()

print('Price Median:', median_price) #mediana es baja que la media, significa que hay outlayers hacia arriba.
```

```python
#range
range_price=diamonds_df['price'].max()-diamonds_df['price'].min()
print('range:', range_price)
```

```python
#standard deviation
std_dev = diamonds_df['price'].std()

print('Standard deviation:', std_dev)
```

```python
diamonds_df.sort_values(by='price')
percentages = [0.05, 0.25, 0.5, 0.75, 0.95]
percentiles = diamonds_df['price'].quantile(percentages)
percentiles
```

```python
# Boxplot (Matplotlib whiskers implementation = 1.5 IQR) for PRICES

data = diamonds_df['price']
ax = data.plot.box(figsize=(6, 6))
ax.set_ylabel('price')
ax.boxplot(data, whis=[0, 100])
plt.tight_layout()
plt.grid()
plt.show()
```

```python
# Histogram ( plotting the frequency table) for PRICES


data = diamonds_df['price']
ax = data.plot.hist(figsize=(12, 8))

ax.set_xlabel('prices')
plt.tight_layout()
```
