```python
df=pd.read_csv('./data/diamonds_data_train.csv', index_col=0)
df.head()
```

```python
df.describe()
df.info()
cols = ['city','cut', 'color', 'clarity']

cat_list = []
for col in cols:
    cat = df[col].unique()
    cat_num = len(cat)
    cat_dict = {"categorical_variable":col,
                "number_of_possible_values":cat_num,
                "values":cat}
    cat_list.append(cat_dict)
    
categories = pd.DataFrame(cat_list).sort_values(by="number_of_possible_values",
                                                ascending=False).reset_index(drop=True)
categories
```

```python
#<one hot encoding:

df_one_hot_encoding = pd.get_dummies(df, 
                                          columns=cols, 
                                          drop_first=True,
                                    )
df_one_hot_encoding.tail()
```

```python
# Model predictions

predictions = model.predict(X_test)

print(type(predictions))
```
